```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

import statistics as sci
# import matplotlib
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
# settings for seaborn plotting style
sns.set(color_codes=True)
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(9.5,5)})
```

```python
sample_size=10000
param_mean = [0, 2]
param_cov = [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(param_mean, param_cov, size=sample_size)
# create a data frame from the data
df = pd.DataFrame(data, columns=["x", "y"])
sns.jointplot(x="x", y="y", data=df, alpha=0.3).plot_joint(sns.kdeplot, zorder=0, n_levels=6)
```

```python
df2 = df.round()
sns.jointplot(x="x", y="y", data=df2, alpha=0.3 , kind = "hist").plot_joint(sns.kdeplot, zorder=0, n_levels=6)
```

```python
from scipy import stats
px = 0.1
py = 1
x = stats.binom.pmf(df2.x,sample_size,px)
y = stats.binom.pmf(df2.y,sample_size,py)
joint_pdf_D = np.outer(x, y)
print(joint_pdf_D)
```

```python
pdf1 = stats.norm.pdf(df.x, sci.mean(df.x),sci.stdev(df.x))
pdf2 = stats.norm.pdf(df.y, sci.mean(df.y),sci.stdev(df.y))
joint_pdf_C = np.outer(pdf1, pdf2)
print(joint_pdf_C)
```
