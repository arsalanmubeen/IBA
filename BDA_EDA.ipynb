```
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
from matplotlib import pyplot as plt
```

```
df= pd.read_csv('/content/sampledata.csv')
df.head()
```

```
df.info()
```

```
df.describe()
```

Show point estimates and confidence intervals as rectangular bars.

```
sns.barplot(x='type',y='amount',data=df)
```

Compute pairwise correlation of columns, excluding NA/null values.

```
numerical_colss=['amount','oldbalanceOrg','newbalanceOrig']
df[numerical_colss].corr()
```

```
sns.heatmap(df[numerical_colss].corr())
```

Plot pairwise relationships in a dataset.

```
sns.pairplot(data=df[df.columns[1:]],diag_kws={'edgecolor':'k','bins':25},plot_kws={'edgecolor':'k'})
plt.show()
```
